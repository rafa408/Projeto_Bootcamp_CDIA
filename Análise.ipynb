```python
import pandas as pd

def analyze_correlations(csv_file):
    """
    Carrega um arquivo CSV, calcula a correlação entre as colunas numéricas
    e exibe as correlações mais fortes em texto.

    Args:
        csv_file (str): O caminho para o arquivo CSV.
    """

    try:
        df = pd.read_csv(csv_file)
    except FileNotFoundError:
        print(f"Erro: Arquivo '{csv_file}' não encontrado.")
        return

    # Selecionar apenas colunas numéricas
    numeric_df = df.select_dtypes(include=['number'])

    # Calcular a matriz de correlação
    corr_matrix = numeric_df.corr()

    # Obter as correlações mais fortes (excluindo a diagonal e duplicatas)
    strong_correlations = (corr_matrix.where((abs(corr_matrix) > 0.5) & (abs(corr_matrix) < 1.0))
                           .stack()
                           .sort_values(ascending=False))

    if strong_correlations.empty:
        print("Não foram encontradas correlações fortes (acima de 0.5).")
    else:
        print("Correlações mais fortes:")
        for (col1, col2), corr in strong_correlations.items():
            print(f"- {col1} e {col2}: {corr:.2f}")  # Formata a correlação para 2 casas decimais


# Exemplo de uso:
analyze_correlations("bootcamp_train.csv")  # Substitua pelo caminho do seu arquivo
```
